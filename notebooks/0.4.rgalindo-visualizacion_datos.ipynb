# Visualizaci√≥n de Datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Diagramas de Dispersion (scatterplot)

x = np.arange(2, 20)
y = x**2
y2 = x**3

plt.scatter(x, y, color='black', marker='*', label='x^2')
plt.scatter(x, y2, color='green', marker='s', label='x^3')
plt.plot(x, y2, color='pink', label='line')

plt.title('Funciones Polin√≥micas', fontsize=20, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()

plt.text(5, 4000, 'texto', color='blue')
```

```python
# Histogramas
mean = 5
std = 3
n = 10000
x = np.random.normal(mean, std, n)

plt.hist(x, bins=20);

plt.title('Histograma de una variable aleatoria', fontsize=20, color='red')
plt.xlabel('x')
plt.ylabel('frecuencia')
plt.grid()
plt.legend()

plt.text(6, 1400, f'Mean: {mean} +/- {std}', color='blue')
```

```python
plt.boxplot(x, vert=False);
```

```python
# Subplots
fig, ax = plt.subplots(ncols=1, nrows=2, figsize=(5, 10), sharex=True)

ax[0].hist(x);
ax[1].boxplot(x, vert=False);
```

```python
# Barplot

classes = ['gatos', 'perros', 'ratones']
totales = [20, 35, 12]

plt.barh(classes, totales)
```

```python
import seaborn as sns

data = pd.DataFrame({
    'Animales': classes,
    'Totales': totales
})

#data.head()

plots = sns.barplot(x='Animales', y='Totales', data=data)

for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.1f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=15, xytext=(0, 8),
                   textcoords='offset points')
```
